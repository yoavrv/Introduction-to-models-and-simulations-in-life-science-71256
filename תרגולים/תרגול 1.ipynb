```python
rng = np.random.default_rng(seed=1452)
```

<div style='direction: rtl'>
    המספרים שהגנרטור מוציא לא באמת אקראיים: הגנרטור משתמש בערך הגרעין ובמצב פנימי ניסתר. כל פעם שאנחנו מבקשים מספר אקראי, יש אלגוריתם ש"מערבב" את הגנרטור והמצב הניסתר כדי ליצור מספר חדש ומצב חדש באופן שהמספר מתפלג באופן שהוא סטטיסטית אחיד.<br>
</div>

```python
rng = np.random.default_rng(seed=1452)
print(rng, ' מצב ראשוני של גנרטור ')
print(rng.bit_generator.state['state'])  # initial state
random_numbers_so_far = []
states_so_far = []
```

```python
random_num = rng.random() # get random number
print('מספר פסאודו רנדומלי חדש:', random_num)
print( rng.bit_generator.state['state'], ':מצב גנרטור')
print('-'*50)
random_numbers_so_far.append(random_num) # save random number and c
states_so_far.append(rng.bit_generator.state['state']['state']) # save rng state
for r, i  in zip(random_numbers_so_far, states_so_far):
    print(r, i)
```
