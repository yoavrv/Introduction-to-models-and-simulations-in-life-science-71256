<h1 style="direction: rtl"><b> תרגול 6: סלקציה דיפלואידית, המשוואה האפינית ל-mRNA </b></h1>

<h2 style="direction: rtl"> בעיה 1</h2>

<div style="direction: rtl"> Problem 3.14 from the book </div>

<div style="direction: rtl"> באופן כללי, זיווג באוכלוסיה הוא לא רנדומלי. אם האוכלוסייה מפוצלת גיאוגרפית, או שיש העדפה לזיווג עם פרטים עם קירבה גנטית, אז בזיווג גמטות שמכילות סטים של גנים דומים יהיו יותר נפוצות מהצפוי בזיווג רנדומלי. זה קורה בגלל שלהורים יש קירבה משפחתית (in-breeding) ולכן הגנים שלהם דומים. </div>

<div style="direction: rtl"> הדרך הפשוטה ביותר להכניס את זה למודל להוסיף פרמטר $f$ שמסמן את הסיכוי להכלאת קרובים (וסיכוי $1-f$ להכלאה רנדומלית). מזה יוצא שאם האחוז באוכלוסיה של אלל $A$ הוא $p\left(t\right)$, אז הסיכוי לזיגוטה מסוג $AA$ הוא $$ p_{AA} = \underbrace{ \left(1-f\right) p\left(t\right)^2}_{\text{random breeding}} + \underbrace{f p\left(t\right)}_{\text{in-breeding}} $$ 
</div>

<div style="direction: rtl"> מה קרה פה? יש לנו שני מקרים: <br>
    יש סיכוי $1-f$ להזדווגות רנדומלית, ואז יש סיכוי $p$ שהורה אחד הוא $A$ וסיכוי $p$ שההורה השני הוא גם $A$. <br>
    יש סיכוי $f$  להזדווגות קרובים, ואז יש סיכוי $p$ שההורה הראשון הוא $A$, ו-$100\%$ סיכוי שההורה השני הוא $A$ (כי בגלל הקרבה המשפחתית יש לו אותם גנים).
    

<div style="direction: rtl"> א. כתבו את הסיכויים לזיגוטה מסוג $Aa$ ו-$aa$. <br> ב. הסבירות לזיגוטות הטרוזיגוטיות שונה מאשר מה שחוקי הרדי-ווינברג אומרים. קחו את התוצאות מחלק א. ויישמו בדיאגרמה כמו בחוקי הרדי ווינברג (עמוד 14 במצגת): </div> 

<div style="direction: rtl"> ג. לפי הטבלה החדשה, יצרו את משוואת הרקורסיה ל-$p\left(t+1\right)$ למודל שלנו. <br> ד. בעזרת הרקורסיה, הראו שהסתברות האללים באוכלוסיה לא משתנה אם $W_{AA} = W_{aA} = W_{aa}$ </div>

<h2 style="direction: rtl"> תשובה 1</h2>

<div style="direction: rtl"> <b>א.</b> הסיכוי לזיגוטה מסוג $aa$ היא סימטרית ל-$AA$: יש לנו סיכוי $\left(1-p\right)$ לאלל $a$, ואנחנו יכולים פשוט להחליף את המשתנה במשוואה ל-$AA$:
    $$ p_{aa} = \underbrace{ \left(1-f\right) \left(1-p\left(t\right)\right)^2}_{\text{random breeding}} + \underbrace{f \left(1-p\left(t\right)\right)}_{\text{in-breeding}} $$
    <br> יש סיכוי $1-f$ להזדווגות רנדומלית, ואז יש סיכוי $1-p$ שהורה אחד הוא $a$ וסיכוי $1-p$ שההורה השני הוא גם $a$. <br>
    יש סיכוי $f$  להזדווגות קרובים, ואז יש סיכוי $1-p$ שההורה הראשון הוא $a$, ו-$100\%$ סיכוי שההורה השני הוא $a$ (כי בגלל הקרבה המשפחתית יש לו אותם גנים).
    </div>

<div style="direction: rtl"> יש לנו שתי דרכים להשיג את הסיכוי לזיגוטה $Aa$: או לחשב בתוך הבנה, או פשוט להשתמש בזה שהסיכויים סוכמים ל-$1$, ולחשב מתוך $p_{aa}+p_{AA}+p_{Aa}=1$. <br>
    מתוך הבנה, נסתכל על המקרים: <br>
    הזדווגות רנדומלית בסיכוי $1-f$, ואז יש סיכוי $p$ שההורה הראשון הוא $A$ ו-$1-p$ שההורה השני הוא $a$.<br>
    הזדווגות רנדומלית בסיכוי $1-f$, ואז יש סיכוי $1-p$ שההורה הראשון הוא $a$ ו-$p$ שההורה השני הוא $A$.<br>
    הזדווגות קרובים בסיכוי $f$, אין סיכוי בכלל $0$ שהורה אחד שונה מהשני.
    $$ p_{aA} = \underbrace{ \left(1-f\right) \left[ p\left(t\right) \left(1-p\left(t\right)\right) + \left(1-p\left(t\right)p\left(t\right)\right)\right]}_{\text{random breeding}} + \underbrace{f\; 0}_{\text{in-breeding}} $$
    </div>

<div style="direction: rtl"> מתוך סכימת סיכויים:
    $$ p_{aA} = 1-p_{aa}-p_{AA} $$
    $$ = 1 - \left(1-f\right) \left[ p^2 + \left(1-p\right)^2 \right] - f\left[p+1-p\right] $$
</div>

<div style="direction: rtl"> נזכור את ההשלמה לריבוע $\left[p + \left(1-p\right)\right]^2 = 1^2 =  p^2 + 2p\left(1-p\right) + \left(1-p\right)^2$  
    </div>

$$ = 1 - \left(1-f\right) \left[ 1-2p\left(1-p\right) \right] - f $$
$$ = 1- 1 + f + 2\left(1-f\right)p\left(1-p\right) -f $$
$$ = 2\left(1-f\right)p\left(1-p\right) $$

<div style="direction: rtl"> נציב את הסיכויים אחרי הסלקציה מהטבלה: נמשיך להשתמש ב-$p\equiv p\left(t\right)$ כי אין מקום

    </div>

    </div>

$$ \overline{W} = \overline{W}p_{AA} + \overline{W}P_{aA}+ \overline{W}P_{aA}++ \overline{W}P_{aa} = p^*_{AA} + 2p^*_{aA}+  p^*_{aa} $$

    </div>

$$ p\left(t+1\right) = p \equiv p \left(t\right) $$

<h2 style="direction: rtl"><b> בעיה 2 </b></h2>

<div style="direction: rtl"> Problem 6.4 from the book </div>

<div style="direction: rtl"> רמת ה-mRNA בתא נקבעת בתהליך ייצור של mRNA ע"י שיעתוק מ-DNA ותהליך התפרקות מולקולות mRNA  קיימות. המודל המתמטי דומה מאוד למודל האוכלוסייה שהשתמשנו בו עד עכשיו: <br> נסמן את קצב הייצור של מולקולות mRNA לשנייה ב-$m$ ואת קצב ההתפרקות העצמית של מולקולות mRNA ב-$r$ מולקולות לשנייה. </div>

<div style="direction: rtl"> במודל הזה, קצב השינוי של רמת ה-mRNA ניתנת על ידי המשוואה הדיפרנציאלית: </div>

$$ \frac{dn}{dt}=-rn+m $$

<div style="direction: rtl"> <b> א. </b> מצאו את נקודת שיווי המשקל של רמת ה-mRNA $\hat{n}$. </div>

<div style="direction: rtl"> <b> ב. </b> Iyer and Struhl (1997) העריכו את זמן מחצית החיים של מולקולות mRNA לגן היסטידין של שמרים his3 כ-$660s$  ע"י חסימת הייצור של תעתיקים חדשים מה שמקבע מלאכותית את $m=0$. <br> השתמשו בפתרון הכללי של משוואות הדעיכה האקספוננציאלית $n\left(t\right)=e^{-rt} n\left(0\right)$ כדי להעריך את $r$. </div>

<div style="direction: rtl"> זמן מחצית החיים $t^*$ אומר $n\left(t^*\right)=\frac{1}{2}n\left(0\right)$ </div>

<div style="direction: rtl"> <b> ג. </b> Iyer and Struhl (1997) גם מצאו שכמות תעתיקי ה-mRNA בשמרים הוא בערך 7 מולקולות בתא. בהנחה שהכמות הזאת היא מצב שיווי משקל, השתמשו בכמות הזאת וב-$r$ שנמצא בסעיף ב. כדי להעריך את $m$. </div>

<div style="direction: rtl"> <b> ד. </b> פתרו את המשוואה באמצעות ע"י הפרדת משתנים. השתמשו בתנאי ההתחלה $n\left(0\right)=n_0$ בשביל למצוא את משתנה האינטגרציה. </div>

<div style="direction: rtl"> <b> ה. </b> בזמן $t=0$, מפעילים מכת חום על השמרים כל שכל תעתיקי ה-$mRNA$ בתא מפורקים. כמה זמן יקח יקח למספר התעתיקים להגיע למחצית מהערך בשיווי משקל $n\left(t\right)=\hat{n}/2$ (כשהתחלנו עם $n\left(0\right)=0$)? האם הזמן הזה תלוי ב-$r$, קצב הפירוק, או $m$, קצב הייצור? </div>

<div style="direction: rtl"> <b> ו. </b> ציירו את הגרף של פתרון המשוואה הכללי באופן תכנותי בעזרת הערכים של $r$ ו-$m$ מהסעיפים הקודמים. בדקו שהגרפים שמתקבלים תואמים לתשובה מסעיף ה. </div>

<h2 style="direction: rtl"> תשובה 1</h2>

<div style="direction: rtl"> <b> א. </b> בנקודת שיווי המשקל, רמת ה-mRNA לא משתנה:. </div>

$$ \frac{dn}{dt}=0=-r\hat{n}+m$$
$$ \hat{n}=\frac{m}{r}$$

<div style="direction: rtl"> <b> ב. </b> נשים את זמן מחצית החיים $t^*$ בתוך המשוואה לדעיכה אקספוננציאלית $n\left(t\right)=e^{-rt} n\left(0\right)$ ונפתור ל-$n\left(t^*\right)=\frac{1}{2}n\left(0\right)$. </div>

$$n\left(t^* = 660\right)=\frac{1}{2}n\left(0\right) =e^{-r\times 660s} n\left(0\right)$$

<div style="direction: rtl"> נחלק ב-$n\left(0\right)$ </div>

$$\frac{1}{2} =e^{-r\times 660s} $$

<div style="direction: rtl"> ניקח $\ln$ </div>

$$ \ln \frac{1}{2} = -\ln2 =-r\times 660s $$

$$ r = \frac{\ln 2}{660s} \approx  \frac{0.693}{660} \frac{1}{s} = 0.00105 \frac{1}{s} = 1.05mHz $$

<div style="direction: rtl"> <b> ג. </b> נציב בביטוי שקיבלנו בסעיף א. את ה-$r$ שקיבלנו בסעיף ב. ונשווה ל-$7$ </div>

$$ \hat{n} = 7 \text{#} = \frac{m}{r} $$
$$ m = 7\text{# }r = 7\text{# }\times 1.05 mHz = 7.35 \text{#}mHz$$

<div style="direction: rtl"> או תעתיק אחד כל $136$ שניות </div>

<div style="direction: rtl"> <b> ד. </b> נכתוב את המשוואה: </div>

$$ \frac{dn}{dt} = -rn+m $$

<div style="direction: rtl"> אין תלות ב-$t$ בצד הימני, אז נחלק בו $-rn+m$ להעביר לצד עם ה-$dn$ ונכפיל את שני הצדדים ב-$dt$. </div>

$$ \frac{1}{-rn+m}dn = dt $$

<div style="direction: rtl"> המשתנים מופרדים: נבצע אינטגרציה: </div>

$$ \int \frac{1}{-rn+m}dn = \int dt $$

$$  = t+A $$

<div style="direction: rtl"> כש-$A$ הוא קבוע האינטגרציה. <br> בשבלי לפתור את האינטגרל, ננסה לעשות החלפת משתנים כך שהמכנה יהיה פשוט ונקווה לטוב: </div>

$$\int \frac{1}{-rn+m}dn$$

$$ u = -rn+m \quad du = -r dn $$

$$ = \int \frac{1}{u}\frac{du}{-r} = -\frac{1}{r} \int \frac{1}{u}du$$

<div style="direction: rtl"> למזלנו, הגענו לאינטגרל ידוע, כפול קבוע פשוט. הפתרון לאינטגרל הוא $\ln$, ונציב חזרה במשוואה: </div>

$$ -\frac{1}{r} \int \frac{1}{u}du = - \frac{1}{r} \ln u= t + A $$

<div style="direction: rtl"> כל פעם שמכפילים או מחלקים את קבוע האינטגרציה, אפשר פשוט לקרוא לדבר החדש קבוע אינטגרציה אחר: </div>

$$ \ln u = -rt -rA = -rt + A^\prime $$

$$  u = \exp\left(-rt + A^\prime \right) = C\exp{} \left(-rt\right)  $$

<div style="direction: rtl"> נציב חזרה את $u=-rn+m$: </div>

$$ -rn + m = C e^{-rt} $$

$$ n + \frac{m}{-r} = C^\prime e^{-rt} $$

$$ n = \frac{m}{r} + C^\prime e^{-rt} $$

<div style="direction: rtl"> נציב בזמן $0$ את רמת ה-mRNA ההתחלתית $n\left(0\right)=n_0$ ונפטר מקבוע האינטגרציה: </div>

$$ n\left(0\right) = \frac{m}{r} + C^\prime e^{0} = n_0 $$

$$  C^\prime  = n_0 - \frac{m}{r}$$

$$ n\left(t\right) = \frac{m}{r} + \left(n_0 - \frac{m}{r} \right)  e^{-rt} $$

$$ n\left(t\right) = \hat{n} + \left(n_0 - \hat{n} \right)  e^{-rt} $$

<div style="direction: rtl"> מה שהמשוואה הזאת אומרת זה שההפרש מהממוצע דועך אקספוננציאלית: </div>

$$ \left(n\left(t\right)-\hat{n}\right) = \left(n_0 - \hat{n} \right)  e^{-rt} $$

$$\delta \left(t\right) = \delta \left(0\right) e^{-rt}$$

<div style="direction: rtl"> <b> ה. </b> במקרה של מכת החום, בזמן ההתחלתי $n\left(0\right)=0$. בשביל למצוא את הזמן בו מגיעים ל-$n\left(t^*\right)=\hat{n}/2$, נציב במשוואה שהשגנו קודם: </div>

$$ \frac{\hat{n}}{2} = \hat{n} + \left(0- \hat{n} \right)  e^{-rt^*} $$

$$ \frac{\hat{n}}{2} = \hat{n} \left(1 -  e^{-rt^*} \right) $$

$$ \frac{1}{2} = 1 -  e^{-rt^*} $$

$$  e^{-rt^*} = \frac{1}{2} $$

$$  rt^* = \ln 2 $$

$$ t^* = \frac{\ln 2}{r} $$

<div style="direction: rtl"> אפשר להציב את $r$ שהשגנו קודם, אבל זה בדיוק המשוואה ממנה השגנו את זמן מחצית החיים: </div>

$$ t^* = \frac{\ln 2}{1.05mHz} = \frac{\ln 2}{\frac{\ln 2}{660s}} = 660s $$

<div style="direction: rtl"> וקיבלנו שהזמן להגיע לחצי מהכמות בשיווי משקל הוא $660$ שניות. אפשר לראות את זה גם ממשוואת דעיכת ההפרש: זה זמן מחצית החיים שההפרש $\hat{n}-0$ מגיע לחצי מהגודל המקורי. </div>

<div style="direction: rtl"> <b> ו. </b> נצייר את הגרף. </div>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_eq = 7
t_half = 660
```

```python
r = np.log(2)/t_half
m = n_eq*r
```

```python
n_0 = 0
t = np.linspace(0, 5*t_half,150)
n = n_eq + (n_0 - n_eq) * np.exp(-r*t)
```

```python
plt.plot(t,n,label=r'$n\left(t\right)$')
r'$n\left(t\right)$'
r'$    $'
plt.axhline(n_eq/2, color='k', linestyle='--',label=r'$\frac{\hat{n}}{2}$')
plt.plot([t_half,t_half],[0, n_eq/2], '--k',label=fr'$t={t_half}$')
plt.title("mRNA levels after heat shock")
plt.xlabel('time (seconds)')
plt.ylabel('mRNA molecules')
plt.legend()
plt.show()
```
