<div dir="rtl"> לדוגמא, יש לנו מידע על כמות המים בעציצים במובא וגוש מידע כזה: </div>

````python
water_level_days_0 = np.array([0,0.1,0.05,0.12])
water_level_nights_0 = np.array([0.1,0.2,0.175,0.13])
water_level_days_1 = np.array([1,1.2,1.3,1.2])
water_level_nights_1 = np.array([1.3,1.4,1.2,1.35])
water_level_days_2 = np.array([0.9,0.7,0.8,0.85])
water_level_nights_2 = np.array([0.95,0.8,0.73,0.88])
````

<div dir="rtl"> ואנחנו רוצים לצייר אותו. יש לנו שלושה עציצים, ואנחנו מודדים את כמות המים ביום ובלילה בנפרד. אנחנו רוצים לצייר שורה של המדידות ביום, ושורה של המדידות בלילה ובכל שורה אנחנו רוצים שלוש עמודות כדי להציג את הגרף של כמות המים בכל עציץ בנפרד
</div>

````python
# נייצר מלבן של צירים
fig, axes = plt.subplots(2,3)
# axes.shape == (2,3)

# נאגד את העציצים ברשימה
water_level_days = (water_level_days_0,
                    water_level_days_1,
                    water_level_days_2)
water_level_nights = (water_level_nights_0,
                       water_level_nights_1,
                       water_level_nights_2)

# נעבור על כל העציצים בלולאה
for i in range(3):
    axe1 = axes[0,i] # נעבור על כל עציץ/עמודה ביום/בשורה העליונה

    axe1.plot(water_level_days[i],'b') # נשתמש בפונקציות של אוביקט הצירים
    axe1.set_title(f'plant #{i}, daytime') # set שימו לב ל
    axe1.set_ylabel('water level')
    axe1.set_ylim(0,2)

    axe2 = axes[1,i] # נעבור על כל עציץ/עמודה בלילה/בשורה התחתונה

    plt.sca(axe2) # ניקח פוקוס לצירים
    plt.plot(water_level_nights[i],'r')
    plt.title(f'plant #{i}, nighttime')
    plt.ylabel('water level')
    plt.ylim(0,2)

plt.tight_layout() # פונקציה שמונעת מהצירים לעלות אחד על השני
plt.show()
````

<div dir="rtl"> מה אפשר לעשות אם אנחנו רוצים לעשות במחשב משהו כמו "השבוע יש 20% סיכוי לגשם", או "יש 2% לקלף נדיר בחבילת קלפים חדשה"? מתמטית, אפשר לבטא את זה עם משתנה אקראי שנקרא ניסיון ברנולי (bernoulli trial) שדומה להטלת מטבע מוטת. המשתנה הרנדומלי הוא <code>True</code> (או 1) בסיכוי p ו-<code>False</code> (או 0) בסיכוי <span dir="ltr">1-p</span>. </div>

$$ \text{Bernoulli trial} \sim \begin{cases} True & p\\ False & q=1-p \end{cases} $$

<div dir="rtl"> אנחנו יכולים להשתמש במספרים אקראיים בשביל לייצר הטלת מטבע כזאת בקוד: נגריל מספר בין 0 ל-1 בעזרת <code>rng.random</code>, ואז נבדוק האם הוא קטן מהסיכוי p. בגלל שההתפלגות היא אחידה בתחום <span dir="ltr">[0,1)</span>, יש סיכוי p שהוא יהיה בתחום <span dir="ltr">[0,p)</span> וסיכוי <span dir="ltr">1-p</span> שהוא יהיה ב-<span dir="ltr">[p,1)</span>  </div>

<div dir="rtl"> אם נריץ את קטע הקוד הבא כמה פעמים, נקבל כל פעם או Head או Tail בסיכוי של 70% </div>

````python
tmp_rng = np.random.default_rng(2220)
````

````python
for i in range(10):
    x=tmp_rng.random()
    is_head = x<0.7
    if is_head:
        print('trial',i,': Head |',f'{is_head=}')
    else:
        print('trial',i,': Tail |',f'{is_head=}')
````

<div dir="rtl"> זה מראה שהביטוי <code dir="ltr">x=rng.random()&lt;p </code> מתאר את הטלת מטבע מוטה, שזה מה שרצינו. </div>
