#### <div dir="rtl"> מעביר עברית בטעות לאנגלית וההפך </div>

```python
engs="`qwertyuiop[]\\asdfghjkl;'zxcvbnm,./~QWERTYUIOP{}|ASDFGHJKL:\"ZXCVBNM<>?"
hebs=";/'קראטוןםפ][\\שדגכעיחלךף,זסבהנמצתץ.~QWERTYUIOP}{|ASDFGHJKL:\"ZXCVBNM><?"
heb_to_eng = {x:y for x,y in zip(engs,hebs)}
eng_to_heb = {y:x for x,y in zip(engs,hebs)}
def is_likely_hebrew(text):
    score_heb = sum((text.count(x) for x in hebs))
    score_eng = sum((text.count(x) for x in engs))
    return score_eng<score_heb

def invert(text):
    if is_likely_hebrew(text):
        return "".join(eng_to_heb.get(x,x) for x in text)
    return "".join(heb_to_eng.get(x,x) for x in text)
```

```python
text="""
nkfatbmhekupsh,u
""".strip()
print(f"is {text} likely written hebrew?",["no", "yes"][is_likely_hebrew(text)])
print("The text is inverted to",invert(text))
```
