<h1 style="direction: rtl"> על הזזה ומתיחה </h1>

<h3 style="direction: rtl"> הזזה של נקודות </h3>

<div style="direction: rtl"> גרף הוא קבוצה של נקודות $\left\lbrace x_i,y_i \right\rbrace$ שמוצגות במרחב</div>

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(0,10,200)
y=x*(np.exp(-x)+x/500) # arbitrary function
plt.plot(x,y,'b')
plt.plot(x,y,'.b')

print(f'points: ({x[0]},{y[0]}), ({x[1]:.4},{y[1]:.4}), ({x[2]:.4},{y[2]:.4}),...')
plt.show()
```

<div style="direction: rtl"> בשביל להזיז את כל הנקודות, נוסיף לכולן הזזה:
$$ \left\lbrace x_i^\prime,y_i^\prime \right\rbrace = \left\lbrace x_i+\Delta x,\; y_i+\Delta y \right\rbrace  $$</div>

```python
dx, dy = -1, 0.2
plt.plot(x,y,'b', label='original graph')
plt.plot(x+dx,y+dy,'g', label='displaced graph')
plt.legend()
plt.show()
```

<div style="direction: rtl"> אפשר להזיז ככה כל צורה, לא רק גרף</div>

```python
# make a polygon
poly_n = 7
# points are equally spaced on a circle: set radius 2, get degrees 0, 360/n ... (1-1/n)360
r, theta = 2, np.linspace(0,2*np.pi,poly_n+1)[:-1]
x_poly, y_poly = r*np.sin(theta), r*np.cos(theta)

# plot a polygon and a displaced polygon
plt.fill(x_poly, y_poly, label='original')
plt.fill(x_poly+dx, y_poly+dy, label=f'displaced by $\\Delta x={dx+0.:.4}$   $\\Delta y ={dy+0.:.4}$')
plt.axis('equal')
plt.legend()
plt.show()
```

<h3 style="direction: rtl"> הזזה של פונקציה </h3>

<div style="direction: rtl"> באופן תיאורטי, ציר ה-$x$ וה-$y$ הם אינסופיים, והדבר שאנחנו עובדים איתו הוא פונקציה $f\left(\right)$ שמקשרת בין $x$ ל-$y$ </div>

<div style="direction: rtl"> בשביל להזיז את גרף הפונקציה, אנחנו צריכים פונקציה חדשה $f^{\prime}$, שהגרף שלה מוסט </div>

<div style="direction: rtl"> אם נקח $y=f\left(x\right)$, אז הגרף הוא אוסף הנקודות $\left\lbrace x,f\left(x\right) \right\rbrace$, אם נעשה הזזה לנקודות- $\left\lbrace x+\Delta x,f\left(x\right)+\Delta y \right\rbrace$ אז נראה שסט הנקודות מתאים לפונקציה חדשה:</div>

<div style="direction: rtl"> $$ f^\prime\left(x+\Delta x \right) = f\left(x\right) + \Delta y $$ </div>

<div style="direction: rtl"> נציב $x$ בתוך הפונקציה החדשה $f^\prime\left(x\right)$ נקבל: </div>

<div style="direction: rtl"> $$ f^\prime\left(x \right) = f\left(x - \Delta x\right) + \Delta y $$ </div>

```python
dx, dy = -1, -0.2
f = lambda x: x*(np.exp(-x)+x/500)
plt.plot(x,f(x),label='$f\\left(x\\right)$')
f_tag = lambda x: f(x-dx)+dy
plt.plot(x,f_tag(x),label='$f^{\\;\\prime}\\left(x\\right)=f\\left(x\\right)+1$')
plt.legend()
plt.show()
```

<h3 style="direction: rtl"> מתיחה של נקודות (scaling) </h3>

<div style="direction: rtl"> כשאנחנו רוצים למתוח גרף, אנחנו רוצים שהמרחק בין כל שתי נקודות בציר שאנחנו מותחים תגדל בפקטור קבוע. אם נnתח בציר ה-$y$ בלבד:</div>

$$ y_i - y_j = \delta y  \xrightarrow{\text{scale}} y^\prime_i - y^\prime_j = B \times \delta y $$

<div style="direction: rtl"> זה יהיה נכון לכל זוג נקודות רק אם <bold>כל</bold> ערכי ה-$y$ של כל הנקודות יוכפלו באותו פקטור. </div>

$$ \left\lbrace x_i ,y_i \right\rbrace \xrightarrow{\text{scale y by }s_y} \left\lbrace x_i ,s_y y_i \right\rbrace $$

```python
s_y = 2
plt.plot(x,y,'b', label='original graph')
plt.plot(x,s_y*y,'g', label='2 $\\times$ scaled graph in y')
plt.legend()
plt.show()
```

<div style="direction: rtl"> מתיחה בציר ה-$x$ באופן דומה:<div>

$$ \left\lbrace x_i ,y_i \right\rbrace  \xrightarrow{\text{scale x by }s_x} \;\xrightarrow{\text{scale y by }s_y} \left\lbrace s_x x_i ,s_y y_i \right\rbrace $$

<div style="direction: rtl"> מתיחה בפקטור גדול מ-$1$ מותחת את הציר, מתיחה בפקטור קטן מ-$1$ מכווצת את הגרף הציר. מתיחה בפקטור שלילי הופכת את הציר<div>

```python
s_x, s_y = 2, 0.2
plt.plot(x,y,'b', label='original graph')
plt.plot(s_x*x,s_y*y,'g', label='scaled graph: $\\times 2$ in x, $\\times 0.2$ in y')
plt.legend()
plt.show()
```

<div style="direction: rtl"> מתיחה היא יחסית לנקודה $\left(0,0\right)$: ככל שהנקודה רחוקה יותר היא זזה רחוק יותר, ולכן צורות שרחוקות מ-$\left(0,0\right)$ יראו מוסטות </div>

```python
# plot a polygon and a displaced polygon
dx, dy = 9,5
s_x, s_y = 1.2, 2
plt.fill(x_poly, y_poly, color='C0', alpha=0.4,label='original')
plt.fill(x_poly+dx, y_poly+dy, color='C1', alpha=0.4,
         # label=f'displaced by $\\Delta x, \\Delta y={dx+0.:.4},{dy+0.:.4}$'
         label='displaced polygone'
        )
# and scale them them
plt.fill(s_x*x_poly, s_y*y_poly, color='c', alpha=0.2,label=f'scaled by $s_x, s_y = {s_x+0.:.4},{s_y+0.:.4}$')
plt.fill(s_x*(x_poly+dx), s_y*(y_poly+dy), color='r', alpha=0.2,
         #label=f'displaced by $\\Delta x,\\Delta y$,\n scaled by $s_x, s_y $')
         label=f'displaced, scaled by \n $s_x, s_y = {s_x+0.:.4},{s_y+0.:.4}$')

plt.axis('equal')
plt.legend(loc='upper left')
plt.show()
```

<div style="direction: rtl"> ולכן צריך לשים לב באיזה סדר עושים הזזה ומתיחה ולתקן בהזזה נוספת </div>

<h3 style="direction: rtl"> מתיחה של פונקציה (scaling) </h3>

<div style="direction: rtl"> כמו שעשינו קודם, נתייחס לגרף של הפונקציה כאוסף הנקודות $\left\lbrace x,f\left(x\right)\right\rbrace$, ונבצע מתיחה </div>

$$\left\lbrace x,f\left(x\right)\right\rbrace \xrightarrow{s_x s_y} \left\lbrace s_x x,s_y f\left(x\right)\right\rbrace$$

<div style="direction: rtl"> אוסף הנקודות החדשות תואם לפונקציה </div>

<div style="direction: rtl"> $$ f^\prime\left(s_x x\right) = s_y f\left(x\right)  $$ </div>

<div style="direction: rtl"> נציב בפונקציה החדשה $x$ ונקבל </div>

<div style="direction: rtl"> $$ f^\prime\left(x\right) = s_y f\left(\frac{x}{s_x}\right)  $$ </div>

```python
s_x, s_y = 0.5, 2
f = lambda x: x*(np.exp(-x)+x/500)
plt.plot(x,f(x),label='$f\\left(x\\right)$')
f_tag = lambda x: s_y*f(x/s_x)
plt.plot(x,f_tag(x),label=f'$f^{{\\;\\prime}}\\left(x\\right)={s_y+0.:.4}\;f\\left(\\frac{{x}}{{{s_x+0.:.4}}}\\right)$')
plt.legend()
plt.show()
```

<h3 style="direction: rtl"> מתיחה והזזה של נקודות </h3>

<div style="direction: rtl"> בדרך כלל קודם מותחים ואז מזיזים: </div>

$$ \left\lbrace x,y \right\rbrace \rightarrow  \left\lbrace s_xx+\Delta x,s_yy+\Delta y \right\rbrace $$

<div style="direction: rtl"> יש שכותבים בצורה של אלגברה ליניארית: </div>

$$ \left\lbrack \begin{matrix} x \\ y \end{matrix} \right\rbrack \rightarrow \left\lbrack \begin{matrix} s_x & 0 \\ 0 & s_y \end{matrix} \right\rbrack \left\lbrack \begin{matrix} x \\ y \end{matrix} \right\rbrack + \left\lbrack \begin{matrix} \Delta x \\ \Delta y \end{matrix}  \right\rbrack$$

```python
s_x, s_y = 2, 0.2
dx, dy = 1,0.2
plt.plot(x,y,'b', label='original graph')
plt.plot(s_x*x+dx,s_y*y+dy,'g', label='scaled and shifted graph')
plt.legend()
plt.show()
```

```python
plt.fill(x_poly, y_poly, color='C0', alpha=0.4,label='original')
plt.fill(s_x*(x_poly+dx), s_y*(y_poly+dy), color='r', alpha=0.2,
         label=f'displaced by $\\Delta x,\\Delta y$,\n scaled by $s_x, s_y {s_x+0.:.4},{s_y+0.:.4}$')

plt.axis('equal')
plt.legend(loc='upper left')
plt.show()
```

<h4 style="direction: rtl"> לפונקציות </h4>

<div style="direction: rtl"> ניקח את הנקודות: </div>

$$ \left\lbrace x,y \right\rbrace \rightarrow  \left\lbrace s_xx+\Delta x,s_yy+\Delta y \right\rbrace $$

<div style="direction: rtl"> ונתאים פונקציה לנקודות החדשות $\left\lbrace x,f^\prime\left(x\right)\right\rbrace$<div>

$$ f^{\;\prime}\left(x\right) = s_y f\left(\frac{x-\Delta x}{s_x}\right)+\Delta y $$

<div style="direction: rtl"><a href="https://www.desmos.com/calculator/apys0pgv00">אינטראקטיבי</a></div>

```python
dx, dy = 0.2, 0.3
s_x, s_y = 1.5, 0.8
f = lambda x: x*(np.exp(-x)+x/500)
plt.plot(x,f(x),label='$f\\left(x\\right)$')
f_tag = lambda x: s_y*f((x-dx)/s_x)+dy
plt.plot(x,f_tag(x),label=f'$f^{{\\;\\prime}}\\left(x\\right)={s_y+0.:.4}\;f\\left(\\frac{{\\left(x-{dx+0.:.4}\\right)}}{{{s_x+0.:.4}}}\\right){dy:+.4}$')
plt.legend()
plt.show()
```
